###### It happens all the time: someone gives you data containing malformed strings,
Python, lists and missing data. How do you tidy it up so you can get on with the
analysis?
Take this monstrosity as the DataFrame to use in the following puzzles:

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',

'Budapest_PaRis', 'Brussels_londOn'],

'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],

'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],

'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
    
'12. Air France', '"Swiss Air"']})

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
                               'londON_StockhOlm','Budapest_PaRis', 
                               'Brussels_londOn'],
                    'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
                    'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
                    'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
                    '12. Air France', '"Swiss Air"']})
```

```python
df.head()
```

```python
df['FlightNumber']
```

```python
new_index = np.arange(1,df.From_To.count()+1)
new_index
df.set_index(new_index,inplace=True)
```

```python
for i in np.arange(1,df.From_To.count()+1):
    if pd.isnull(df.FlightNumber.loc[i,]):
        df.loc[i,'FlightNumber'] = df.FlightNumber.loc[i-1,]+10
df['FlightNumber']
```

```python
df['FlightNumber'].astype('int')
```

```python
df['From_To']
```

```python
temporarydf = df.copy()
```

```python
temporarydf
```

```python
temporarydf[['From','To']] = temporarydf.From_To.str.split('_',expand = True)
```

```python
temporarydf
```

```python
temporarydf[['From','To']]
```

```python
temporarydf.From = temporarydf.From.str.capitalize()
temporarydf.To = temporarydf.To.str.capitalize()
temporarydf.From_To = temporarydf.From_To.str.capitalize()
```

```python
temporarydf
```

```python
temporarydf
```

```python
df
```

```python
df.drop('From_To',axis=1,inplace=True)
df
```

```python
df['From_To'] = temporarydf['From_To']
df
```

```python
df_delays = pd.DataFrame(df['RecentDelays'].values.tolist())
df_delays.columns = ['delays_{}'.format(n) for n in range(1,len(df_delays.columns)+1)]
df_delays
```

```python
df.drop('RecentDelays',axis=1)
df = df.join(df_delays)
```
